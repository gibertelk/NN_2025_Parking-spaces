```python
import torch
import pandas as pd

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# сравнение моделей
checkpoint_1 = torch.load('best_faster_rcnn.pth', map_location=device)
checkpoint_2 = torch.load('best_yolo_13.pth', map_location=device)
checkpoint_3 = torch.load('best_yolo_20.pth', map_location=device)

# print(checkpoint_1)


df=pd.DataFrame(columns=["epoch", "mAP50", "valid_loss", "train_loss"])
df.loc['Faster R-CNN']=[f"{checkpoint_1['epoch']}", f"{checkpoint_1['map50']:.4f}", f"{checkpoint_1['valid_loss']:.4f}", f"{checkpoint_1['train_loss']:.4f}"]
df.loc['Yolo 13']=[f"{checkpoint_2['epoch']}", f"{checkpoint_2['map50']:.4f}", f"{checkpoint_2['valid_loss']/42:.4f}", f"{checkpoint_2['train_loss']/377:.4f}"]
df.loc['Yolo 20']=[f"{checkpoint_3['epoch']}", f"{checkpoint_3['map50']:.4f}", f"{checkpoint_3['valid_loss']/42:.4f}", f"{checkpoint_3['train_loss']/377:.4f}"]
print(df)
```
